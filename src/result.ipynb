```python
import matplotlib.pyplot as plt
import numpy as np

model_names = [
    "XGBoost",
    "SVM (full)",
    "SVM (missing)",
    "LogReg (full)",
    "LogReg (missing)",
    "Decision Tree",
    "KNN (full)",
    "KNN (missing)",
]

mean_f1 = np.array([
    0.3380,  # XGBoost
    0.2803,  # SVM (full)
    0.2636,  # SVM (missing)
    0.2483,  # LogReg (full)
    0.2769,  # LogReg (missing)
    0.2703,  # Decision Tree
    0.1584,  # KNN (full)
    0.1596,  # KNN (missing) 
])

std_f1 = np.array([
    0.0063,  # XGBoost
    0.0037,  # SVM (full)
    0.0340,  # SVM (missing)
    0.0030,  # LogReg (full)
    0.0243,  # LogReg (missing)
    0.0080,  # Decision Tree
    0.0095,  # KNN (full)
    0.0121,  # KNN (missing)
])

baseline_f1 = 0.1485 
x = np.arange(len(model_names))

plt.figure(figsize=(7, 4))
bars = plt.bar(x, mean_f1, yerr=std_f1, capsize=4)
plt.axhline(baseline_f1, linestyle="--", linewidth=1, label="Baseline F1 (predict all 1)")


for i, v in enumerate(mean_f1):
    plt.text(
        i,
        v + 0.005,        
        f"{v:.3f}",        
        ha="center",
        va="bottom",
        fontsize=8,
    )

plt.xticks(x, model_names, rotation=30, ha="right")
plt.ylabel("Test F1 score")
plt.title("Model performance vs baseline (F1, mean  std over 3 random states)")
plt.legend()
plt.tight_layout()
plt.savefig("../figures/model_f1_comparison.png", dpi=300, bbox_inches="tight")
plt.show()
```
